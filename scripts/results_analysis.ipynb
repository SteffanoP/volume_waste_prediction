# Analysis of the Result

## Importing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./models/results.csv')
```

```python
df[['Algorithm', 'Type']] = df['Algorithm'].str.split('_', expand=True)
```

```python
df['Algorithm'] = df['Algorithm'].str.replace('Regressor', '')
df['Algorithm'] = df['Algorithm'].str.replace('HistGradientBoosting', 'HistogramGB')
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt

Algorithms = df['Algorithm'].unique()
Types = {
    "Raw": df[df['Type'] == "raw"]['Mean Training time'].to_numpy(),
    "Pre-Processed": df[df['Type'] == "preprocessed"]['Mean Training time'].to_numpy()
}

x = np.arange(len(Algorithms))
width = 0.3
multiplier = 0

fig, ax = plt.subplots()

for attribute, measurement in Types.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, fmt='%.1f', padding=3)  # Format each data point with two decimal places
    multiplier += 1
    
ax.set_xlabel('Algorithm')
ax.set_ylabel('Mean Training Time')
ax.set_xticks(x + (width/2), Algorithms)
ax.set_title('Mean Training Time by Algorithm and Type')
ax.legend(title='Type')

ax.set_ylim(bottom=0, top=ax.get_ylim()[1] * 1.2)
```

```python
import matplotlib.pyplot as plt

Algorithms = df['Algorithm'].unique()
Types = {
    "Raw": df[df['Type'] == "raw"]['Mean R2 score'].to_numpy(),
    "Pre-Processed": df[df['Type'] == "preprocessed"]['Mean R2 score'].to_numpy()
}

x = np.arange(len(Algorithms))
width = 0.35
multiplier = 0

fig, ax = plt.subplots()

for attribute, measurement in Types.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, fmt='%.4f', padding=5)  # Format each data point with four decimal places
    multiplier += 1
    
ax.set_xlabel('Algorithm')
ax.set_ylabel('Mean R2 Score')
ax.set_xticks(x + (width/2), Algorithms)
ax.set_title('Mean R2 Score by Algorithm and Type')
ax.legend(title='Type')

# Increase the y-axis scale
ax.set_ylim(bottom=0, top=ax.get_ylim()[1] * 1.3)
```
