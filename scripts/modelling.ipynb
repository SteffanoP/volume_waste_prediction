```python
import time
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
dataset = pd.read_csv('../data/preprocessed.csv', index_col='PES_ID')
```

```python
dataset['PES_DATAINI'] = (pd.to_datetime(dataset['PES_DATAINI'])).apply(lambda x: time.mktime(x.timetuple()))
dataset['PES_DATAFIM'] = pd.to_datetime(dataset['PES_DATAFIM']).apply(lambda x: time.mktime(x.timetuple()))
dataset['PES_HRINI'] = pd.to_datetime(dataset['PES_HRINI']).apply(lambda x: time.mktime(x.timetuple()))
dataset['PES_HRFIM'] = pd.to_datetime(dataset['PES_HRFIM']).apply(lambda x: time.mktime(x.timetuple()))
```

```python
dataset.head()
```

```python
target_column_name = 'PES_PESOUTIL'
target_column = dataset.pop(target_column_name)
```

```python
X = dataset.to_numpy()
y = target_column.to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
params = {
    "max_depth": 1000,
    "learning_rate": 0.50,
    "loss": "squared_error",
}
```

```python
reg = HistGradientBoostingRegressor(**params, categorical_features="from_dtype")
reg.fit(X_train, y_train)
```

```python
mse = mean_squared_error(y_test, reg.predict(X_test))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```
